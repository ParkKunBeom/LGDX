```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic') #한글 깨짐 방지
```

```python
df = pd.read_csv('./2015-2017이혼사유.csv', encoding='utf-8', index_col='연도') #데이터 불러오기
```

```python
df.loc[:, '계':] = df.loc[:, '계':].replace({',': ''}, regex=True).apply(pd.to_numeric) #숫자형 변환
ratio_cols = df.columns[df.columns.get_loc('계') + 1:] # 각 열을 퍼센테이지 변환을 위한 컬럼 목록 지정
df[ratio_cols] = df[ratio_cols].div(df['계'], axis=0) * 100 # 이혼 사유 전체를 퍼센트 변환
df
```

```python
df_2015 = df.loc[2015].set_index('연령별') # 2015에 해댕하는 모든 행

df_2015.loc[:, '배우자 부정':].plot(kind='barh') # 기존 인덱스를 연령별로 변경 -> y축으로 사용
plt.xlabel('비율(%)')
plt.ylabel('연령별')
plt.title('2015년 연령별 이혼사유 비율')
plt.legend(bbox_to_anchor=(1.5, 0.5))
plt.show()
```
