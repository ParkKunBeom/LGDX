```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic') #한글 깨짐 방지
```

```python
df = pd.read_csv('./data/이혼에 대한 인식.csv', encoding='utf-8', index_col='인식') #데이터 불러오기
```

```python
df = df.replace({'-': ''}, regex=True).apply(pd.to_numeric).fillna(0).drop('합계')
```

```python
df.T.plot(kind='bar',
    stacked=True,
    figsize=(12, 6),
    title='이혼에 대한 인식',
    ylim=(0,100))
plt.xticks(rotation=0)
plt.legend(bbox_to_anchor=(1.5, 0.5))
```
