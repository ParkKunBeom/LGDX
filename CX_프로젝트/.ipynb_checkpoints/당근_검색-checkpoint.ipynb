```python
import time
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from tqdm import tqdm


# 1. 검색어 입력
keyword = input("검색할 키워드를 입력하세요: ").strip()

# 2. 지역 ID 범위
regions = list(range(6, 106))

# 결과 저장 리스트
results = []

# 3. 드라이버 실행
options = Options()
options.add_argument("--start-maximized")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

# 4. 중복 제거
seen = set()

print("상품명|가격")

for region in tqdm(regions, desc="지역 진행상황"):
    url = f"https://www.daangn.com/kr/buy-sell/?in={region}&search={keyword}"
    driver.get(url)
    time.sleep(1)

    spans = driver.find_elements(By.TAG_NAME, "span")

    for s in spans:
        try:
            txt = s.text.strip()

            if not txt:
                continue
            if ("원" not in txt) and ("나눔" not in txt):
                continue

            parent = s.find_element(By.XPATH, "..")
            spans_in_parent = parent.find_elements(By.TAG_NAME, "span")
            texts = [x.text.strip() for x in spans_in_parent if x.text.strip()]

            if len(texts) < 2:
                continue

            name = texts[0]

            real_price = ""
            for t in texts:
                if "원" in t or "나눔" in t:
                    real_price = t
                    break

            if not name or not real_price:
                continue

            key = (name, real_price)
            if key in seen:
                continue
            seen.add(key)

            

            # 결과 저장
            results.append([name, real_price])

        except:
            continue

driver.quit()


# CSV 저장 (한글 깨짐 방지)
filename = f"./data/당근_{keyword}.csv"

with open(filename, "w", newline="", encoding="utf-8-sig") as f:
    writer = csv.writer(f)
    writer.writerow(["상품명", "가격"])
    writer.writerows(results)

print(f"\nCSV 저장 완료 → {filename}")
```
