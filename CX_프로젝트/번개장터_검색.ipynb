```python
import os
import csv
import re
import time
from urllib.parse import quote
from tqdm import tqdm

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait


keyword = input("검색할 키워드를 입력하세요: ").strip()
results = []

options = Options()
# options.add_argument("--headless=new")   # 처음엔 주석 상태로 테스트 권장
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")
options.page_load_strategy = "eager"

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

seen = set()
page = 1
price_pattern = re.compile(r'^\d{1,3}(,\d{3})*$')
encoded_keyword = quote(keyword)

while True:
    url = f"https://m.bunjang.co.kr/search/products?order=score&page={page}&q={encoded_keyword}"
    driver.get(url)

    try:
        WebDriverWait(driver, 10).until(
            lambda d: len(d.find_elements(By.CSS_SELECTOR, 'a[data-pid]')) > 0
        )
    except:
        print(f"{page}페이지: 상품 카드 없음 → 종료")
        break

    cards = driver.find_elements(By.CSS_SELECTOR, 'a[data-pid]')
    print(f"{page}페이지 카드 수: {len(cards)}")

    if len(cards) == 0:
        print(f"{page}페이지: 카드 0개 → 종료")
        break

    page_new_count = 0

    for card in tqdm(cards, desc=f"{page}페이지 수집중", leave=False):
        try:
            text = card.text.strip()
            lines = [line.strip() for line in text.split("\n") if line.strip()]

            if not lines:
                continue

            name = ""
            price = ""

            for i, line in enumerate(lines):
                # 번개장터 가격은 "250,000" 형태
                if price_pattern.match(line):
                    price = line
                    if i > 0:
                        name = lines[i - 1]
                    break

            if not name or not price:
                continue

            if name in ["배송비포함", "AD", "광고"]:
                continue

            key = (name, price)
            if key in seen:
                continue

            seen.add(key)
            results.append([name, price])
            page_new_count += 1

        except:
            continue

    print(f"{page}페이지 신규 수집: {page_new_count}개")

    # 새로 수집된 게 없으면 다음 페이지도 볼 필요가 거의 없으므로 종료
    if page_new_count == 0:
        print("신규 데이터 없음 → 종료")
        break

    page += 1
    time.sleep(1)

driver.quit()

os.makedirs("./data", exist_ok=True)
filename = f"./data/번개장터_{keyword}.csv"

with open(filename, "w", newline="", encoding="utf-8-sig") as f:
    writer = csv.writer(f)
    writer.writerow(["상품명", "가격"])
    writer.writerows(results)

print(f"\nCSV 저장 완료 → {filename}")
print(f"총 {len(results)}개 수집")
```
