### 학습목표
- pandas 라이브러리를 마스터할 수 있다!!
- 인덱싱, 슬라이싱, 행추가, 열추가, 연산 등에 대한 모든 정보를 마스터할 수 있다!!

##### 개념정리
- 인덱싱, 슬라이싱
  - 데이터에 범위를 주어 원하는 데이터를 추출
  - 열인덱싱 -> df['컬럼명']
  - 행슬라이싱 -> df[행번위]
  - 행과 열 단위로 추출 -> 인덱서 (loc,iloc)
    - df.loc[행범위,열범위] => 인덱스명
    - do.iloc[행범위,열범위] => 인덱스번호

```python
import pandas as pd
```

```python
population = pd.read_csv('./data/population.csv', index_col = '도시')
population
```

```python
#1. 서울~인천, 2020~2010 데이터를 추출
population.loc['서울':'인천','2020':'2010']
```

```python
population.iloc[0:3,1:4]
```

- 불리언 인덱싱
  - 조건을 활용하여 조건식이 True인 데이터를 출력
  - df[조건]

```python
#2020년에 인구가 2,000,000 이상인 도시를 출력
population[population['2020']>=2000000]
```

```python
#2020년에 인구가 2,000,000 이상인 도시 이름만 출력
population[population['2020']>=2000000].index
```

```python
#2020년에 인구가 2,000,000 이상인 도시dml 2020년도 인구 출력
population[population['2020']>=2000000]['2020']
```

```python
#열추가
#df[추가할 컬럼명]=[데이터]

population['2000']=[0,0,0,0,0]
population
```

```python
# 행추가 -> 순천 행 추가
population.loc['순천'] = ['전라권',0,0,0,0,0]
population
```

```python
# 결측치(NaN)를 원하는 값으로 바꿔주는 함수
# 결측치끼리의 연산은 결측치로 출력되기 때문!!
# df명.fillna(value = 특정값)
population.fillna(value=0, inplace=True)
population
```

### DataFrame 실습

```python
score = pd.read_csv('./data/score.csv', encoding = 'euc-kr', index_col= '과목')
score
```

Q1. 과목별 합계를 구하여 DataFrame에 추가('합계'컬럼 추가)

```python
# 과목별 합계 -> 열방향연산 -> 축설정 axis = 1 
score['합계']=score.sum(axis=1)
```

```python
score
```

Q2. 과목별 평균을 계산하여 '평균'컬럼 추가

```python
score['평균']=score.iloc[:,0:-1].mean(axis=1)
score
```

Q3. 반별 평균을 계산하여 "반평균" 행 추가

```python
score.iloc[:,:-1].mean()
```

```python
score.loc['반평균']=score.iloc[:,:-1].mean()
```

```python
score.fillna(value='-', inplace=True)
```

```python
score
```
