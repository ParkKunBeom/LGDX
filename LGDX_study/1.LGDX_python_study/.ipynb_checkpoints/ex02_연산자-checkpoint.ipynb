### 산술 연산자
 - 더하기, 빼기, 곱하기, 나누기, 나누기(몫), 나머지

```python
# 변수 num1 공간에 숫자 10 저장
# 변수 num2 공간에 숫자 8 저장
num1 = 10
num2 = 8
# 각 변수의 내용 출력해보기!
print(num1)
print(num2)
# 숫자 타입으로 인식할까?
print(type(num1)) # integer
print(type(num2))
```

```python
# 산술연산자
print(num1 + num2)
print(num1 - num2)
print(num1 * num2)
print(num1 / num2)
print(num1 // num2)
print(num1 % num2)
```

- 산술 연산자 문자열 처리

```python
# str1 = "안녕"
str1 = '안녕'
str2 = '하세요'

# 변수 두개 더해서 출력해보기
print(str1 + str2)
```

```python
# 데이터 타입이 달라도 더해질까?
num1 + str1
```

```python
# '10안녕' 더해지게 하고 싶다면?
# int -> str : str(값)
str(num1) + str1
```

```python
# 문자열 곱하기
'*'*1000
```

```python
# input('설명문'): 키보드로 부터 입력 받는 기능(단, 입력 데이터는 문자열로 인식)
num1 = int(input('두자리 숫자를 입력해주세요>>'))
print(type(num1))
```

```python
num1 = int(input('첫번째 숫자를 입력해주세요>>'))
num2 = int(input('두번째 숫자를 입력해주세요>>'))
# print('더하기 결과:', num1+num2) # print 기능 활용 : , 로 추가하는 개념
print(f'더하기 결과 : {num1 + num2}') # 문자열 포매팅 : f문자열 포매팅기능 
print(f'빼기 결과 : {num1 - num2}')
print(f'곱하기 결과 : {num1 * num2}')
result = num1/num2
print(f'나누기 결과 : {num1 / num2:.3f}') # 3번째 자리까지 표현
```

### 지수 연산자
 - 제곱, 세제곱, ~~승

```python
num = 2
power = int(input('승을 입력하세요>'))
print(f'2의 {power}승 :', num ** power)
```

```python
# q. num 값이 항상 2가 아님. 키보드로 부터 입력받아 지수값을 출력하도록 만들어보기!
num = int(input('지수를 입력하세요>'))
power = int(input('승을 입력하세요>'))
print(f'{num}의 {power}승 :', num ** power)
```

### 대입(복합) 연산자
 - 대입 : = (같다는 표현은 : ==)
 - 복합대입 : += , -=, *=, /=

```python
a = 3
b = '신난다!'
print(a, b)
```

```python
# +=
a += 3
# a= a + 3
print(a)
```

```python
# -=
a = 3
a -= 1
# a = a - 1 >> a = 3 - 1
print(a)
```

- 치환 : 두개의 공간에 있는 데이터의 위치를 바꾸는 개념

```python
a = 3
b = 5
print(a, b)
```

```python
a = b
b = a
print(a,b)
```

```python
a = 3
b = 5
a,b = b,a # 치환될까?
print('a : ', a)
print('b : ', b)
```

### 비교연산자
 - 크거나 같은지 등 : >, >=, <, <=
 - 같거나 다르거나 등 : ==, != (같지 않다)
 - 결과값 (불리언 데이터) : True(1) or False(0)

```python
a = 3
b = 7
# a,b = 3,7
print(a > b) # --> False
print(a < b) # --> True
print(a == b) # --> False
print(a != b) # --> True
```

### 논리 연산자
 - not : 원래의 결과를 반전시키는 키워드
 - and, or : 논리 and 논리 , 논리 or 논리
 - and : 두개의 논리가 Ture(1) 해당하면 True를 출력하는 연산자
 - or : 둘 중의 한개라도 True(1) 해당하면 True를 출력하는 연산자

```python
result = a != b #True
not result
```

```python
(3>2) and (5>1)
```

```python
(3>2) and (0>1)
```

```python
(3>2) or (0>1)
```

### 멤버연산자
 - in : 찾고자 하는 값이 포함되어 있으면 Ture
 - not in : 찾고자 하는 값이 포함되어 있지 않으면 True

```python
str1 = "오늘 점심 뭐 먹어?"
"?" in str1
```

```python
"?" not in str1
```

```python
"!" not in str1
```

```python
# 리스트 : ['딸기', '수박', '참외']
'참외' in ['딸기', '수박', '참외']
```

```python
'멜론' in ['딸기', '수박', '참외']
'멜론' not in ['딸기', '수박', '참외']
```

### 삼항연산자
 - a if 논리 else b
 - True if 논리 else False

```python
# 파이썬에서 if 조건 : 소문자
# 불리언 값 : True, False
True
# true # 단순 변수의 개념
```

```python
# 60점 이상이면 합격
# num 변수에 사용자로부터 시험 점수를 입력받자
num = int(input('시험 점수 >'))
num
# 삼항연산자를 통해서 60이상이면 합격, 미만이면 불합격 문자열을 출력!
result = "합격" if num>60 else "불합격"
print(f'{result} 하셨습니다.')
```

```python
a = int(input("정수입력 >> "))
b = int(input("정수입력 >> "))
result = f'a: {a}' if a>b else f'b: {b}'
print(result)
```

```python
ball = int(input('농구공 수 입력 >> '))

box = f'필요한 상자의 수 : {ball//5}박스' if ball%5 == 0 else f'필요한 상자의 수 : {ball//5+1}박스'
print(box)
```

```python
ball = int(input('농구공 수 입력 >> '))

box = ball//5 if ball%5 == 0 else ball//5+1
print(f'필요한 상자의 수 : {box}박스')
```

```python
worktime = int(input('근무시간 입력 >> '))
money = 10000*worktime  if worktime <= 8 else 10000*8 + 15000*(worktime-8)
print(f'총 지급액은 {money}원 입니다.')
```

```python
db = int(input('DB >> '))
ec = int(input('전자계산기 구조 >> '))
os = int(input('운영체제 >> '))
dc = int(input('데이터통신 >> '))
sw = int(input('소프트웨어공학 >> '))
total = db+ec+os+dc+sw
result= '합격입니다!' if total>=60 and db>=8 and ec>=8 and os>=8 and dc>=8 and sw>=8 else '불합격입니다!'
print(result)
```
