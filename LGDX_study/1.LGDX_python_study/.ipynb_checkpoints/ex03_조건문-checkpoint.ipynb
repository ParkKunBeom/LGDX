## 조건문
 - 상황에 따라 실행 흐름을 다르게 하는 문법

```python
# 각 실행문이 동일한 들여쓰기 위치에 있어야 함
if True:
    print('실행문장 1 실행')
        print('실행문장 2 실행')
print('조건문 종료')
```

```python
if False:
    print('실행문장 1')
elif False:
    print('실행문장 2')
elif False:
    print('실행문장 3')
# else:
#     print('실행문장 4')
print('if문 종료')
```

```python
# num1 과 num2에 숫자를 입력받아 크기를 비교
# 첫 번째 숫자가 더 큽니다. 두 번째 숫자가 더 큽니다. 두 수가 똑같습니다.

num1=int(input('첫 번째 숫자 입력 : '))
num2=int(input('두 번째 숫자 입력 : '))

if num1 > num2 :
    print('첫 번째 숫자가 더 큽니다.')
elif num1 < num2 :
    print('두 번재 숫자가 더 큽니다.')
# elif num1==num2 :
#     print('두 수가 똑같습니다.')
else:
    print('두 수가 똑같습니다.')
```

```python
score = int(input('점수 입력 >> '))

if 100>=score>=90:
    print(f'{score}점은 A학점 입니다.')
elif score>=80:
    print(f'{score}점은 B학점 입니다.')
elif score>=70:
    print(f'{score}점은 C학점 입니다.')
elif score>=60:
    print(f'{score}점은 D학점 입니다.')
else:
    print(f'{score}점은 F학점 입니다.')
```

```python
score = int(input('점수 입력 >> '))

if 100>=score>=90:
    grade='A'
elif score>=80:
    grade='B'
elif score>=70:
    grade='C'
elif score>=60:
    grade='D'
else:
    grade='F'
print(f'{score}점은 {grade}학점 입니다.')
```

```python
print("-----[SM Coffee]-----")
price = int(input('금액을 입력하세요 >> '))
print("커피를 고르세요.")
coffee = int(input('[1]에스프레소(600원) [2]아메리카노(1000원) [3]카페라떼(1300원) >> '))

if coffee==1:
    if price>600:
        print(f'거스름돈: {price-600}원')
    else:
        print('돈이 부족해요..T^T')
elif coffee==2:
    if price>1000:
        print(f'거스름돈: {price-1000}원')
    else:
        print('돈이 부족해요..T^T')
elif coffee==3:
    if price>1300:
        print(f'거스름돈: {price-1300}원')
    else:
        print('돈이 부족해요..T^T')
else:
    print('메뉴잘못입력함')
        
```

```python
inputMoney = int(input("금액을 입려하세요 >> "))
print('커피를 고르세요.')
menu = int(input('[1]에스프레소(600원) [2]아메리카노(1000원) [3]카페라떼(1300원) >> '))
price=0

if menu == 1 :
    price = 600
elif menu == 2 :
    price = 1000
elif menu == 3 :
    price = 1300
else:
    price = 0

if price == 0:
    print('메뉴를 잘 못 입력했어요')
elif inputMoney >= price:
    print(f'거스름돈 : {inputMoney-price}원')
else:
    print('돈이 부족해요ㅜㅜ')
```

```python
inputMoney = int(input("금액을 입려하세요 >> "))
print('커피를 고르세요.')
menu = int(input('[1]에스프레소(600원) [2]아메리카노(1000원) [3]카페라떼(1300원) >> '))

if menu == 1 :
    price = 600
elif menu == 2 :
    price = 1000
elif menu == 3 :
    price = 1300
else:
    price = 0

if price == 0:
    print('메뉴를 잘 못 입력했어요')
elif inputMoney >= price:
    change = inputMoney-price
    print(f'거스름돈 : {change}원')
    print(f'천원:{change//1000}장')
    print(f'오백원:{change%1000//500}개')
    print(f'백원:{change%500//100}개')
else:
    print('돈이 부족해요ㅜㅜ')
```

```python
password = input('비밀번호를 입려하세요 >> ')
if len(password) >= 8:
    password2 = input('비밀번호를 한번 더 입려하세요 >> ')
    if password == password2:
        print('비밀번호가 일치합니다')
    else:
        print('비밀번호가 일치하지 않습니다. 다시 시작해주세요.')    
else:
    print('8자 이상으로 입력 해야합니다. 다시 시작해주세요.')    
```

```python
pw = input("비밀번호를 입력핫요 >> ")

# 8글자인지 판별 조건
if len(pw)<8:
    print('8자 이상으로 입력해야합니다. 다시 시작해주세요.')
else:
    pw_check = input('비밀번호를 한번 더 입력하세요 >> ')

    if pw == pw_check:
        print('비밀번호가 일치합니다.')
    else:
        print('비밀번호가 일치하지 않습니다. 다시 시작해주세요.')
```
