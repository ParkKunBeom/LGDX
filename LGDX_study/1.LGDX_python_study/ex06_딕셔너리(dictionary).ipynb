### 딕셔너리 (dictionary)
- key : value 한쌍으로 이루어진 자료구조
- 순서가 없는 형태 (인덱스 번호 X)
- {} 중괄호를 통하여 생성
- key : 변하지 않는 유일한 값 -> 중복해서 사용했을때 나중값만 인식

```python
student = {'이름' : '김소은', '나이' : '21', '전화번호' : '01012345678'}
student
```

```python
print(type(student))
```

```python
# key 값을 활용해서 value 추출
# 딕셔너리명[key]
student['이름']
```

- value 값 변경이 가능
  - 추가, 수정, 삭제

```python
# 값 추가
# 추가할 값에 대한 새로운 key 값을 대괄호 안에 넣어주고 대입
student['성별'] = '여자'
student
```

```python
# value 수정(나이 21 -> 24 변경)
student['나이'] = 24
student
```

```python
# 값 삭제
# del student[key]
del student['나이']
student
```

- 한개의 key에 여러개의 value 대응

```python
# list로 작성
student['취미']=['코딩','넷플릭스 시청','음악감상']
student
```

- value만 가져오기
  - 딕셔너리명.values()

```python
student.values()
```

- 전체 데이터를 리스트 형식으로 가져오기
  - 딕셔너리명.items()

```python
student.items()
```

```python
for data in student.items():
    for txt in data:
        print(txt)
```

```python
for key, value in student.items():
    print(key, value)
```

- 딕셔너리 멤버연산자 활용
  - key in 딕셔너리명 : 멤버연산자는 딕셔너리의 key에만 동작

```python
'성별' in student
```

- 딕셔너리 값 삭제
  - del 딕셔너리명[key]

```python
del student['성별']
student
```

```python
if '성별' in student:
    del student['성별']
    print('삭제 완료')
else:
    print('해당 key값이 없습니다.')
```

```python
if '전화번호' in student:
    del student['전화번호']
    print('삭제 완료')
else:
    print('해당 key값이 없습니다.')
```
