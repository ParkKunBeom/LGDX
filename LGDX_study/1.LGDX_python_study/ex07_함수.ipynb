- 함수의 구조


```python
# 두개의 숫자를 매개변수로 받아 두 수의 합을 반환하는 함수 정의
def number_sum(n1, n2):
    result = n1+n2
    return result
    print('출력될까?')
```

```python
print(number_sum(5,7))
print(number_sum(8,7))

r = number_sum(20,6)
print(r)
```

```python
# 호출시에는 매개변수의 개수에 맞춰 호출해야함
# number_sum(1)
# number_sum(1,2,3)
number_sum(1,2)
```

```python
# 두 숫자를 매개변수로 받아 더 큰수를 반환, 같을 경우에는 같은 수 반환

def num_max(n1,n2):
    if n1>n2:
        return n1
    else:
        return n2
```

```python
print(num_max(5,9))
print(num_max(10,10))
print(num_max(10,15))
```

- 두 개의 정수와, 한개의 연산기호('+','-')를 매개변수로 받아 연신기호에 맞게 연산하여 결과값을 반환하는 함수 cal 정의

```python
def cal(n1,n2,op):
    if op == '+':
        return n1+n2
    elif op == '-':
        return n1-n2
```

```python
print(cal(5,8,'+'))
print(cal(5,8,'-'))
print(cal(5,8,'/'))
```

- 2개의 정수를 매개변ㅅ로 받아 두 숫자 중 10에 더 가까운 수를 반환하는 close10 함수 정의
- 만약 두 숫자 모두 10과의 차이가 같다면 0을 반환

```python
def close10(n1,n2):
    if 10-abs(n1)<10-abs(n2):
        return n1
    elif 10-abs(n1)>10-abs(n2):
        return n2
    else:
        return 0
```

```python
print(close10(15,-15))
```

```python
def close10(n1, n2):
    temp1 = (10-n1)**2 # -7**2=49
    temp2 = (10-n2)**2 # 5**2=25
        
    if temp1 < temp2:
        return n1
    elif temp1 > temp2:
        return n2
    else:
        return 0
```

```python
print(close10(17,5))
print(close10(17,5))
```

- 두 개의 숫자(start, end)를 매개변수로 받아 start부터 end까지의 숫자를 list로 만들어 반환하는 rangeList 함수 정의

```python
def rangeList(start, end):
    lst=[]
    for i in range(start,end+1):
        lst.append(i)
    return lst
```

```python
rangeList(1,15)
```

```python
def rangeList(start, end):
    return list(range(start,end+1))

rangeList(1,10)
```

- 두수와 연산 기호(+, -)를 사용자에게 입력받아 원하는 연산을 진행하는 함수를 정의, 호출하여 사용

```python
def cal(n1,n2,op):
    """함수 설명: 기호를 입력받아 덧셈과 뺄셈을 계산하는 함수"""
    if op == '+':
        return n1+n2
    elif op == '-':
        return n1-n2
    else:
        return '기호를 잘못입력하셨습니다'
```

```python
# num1, num2, op 사용자에게 입력받아
# cal() 함수를 정의하여 연산할 수 있도록 작성
num1 = int(input('첫 번재 정수 입력 >> '))
num2 = int(input('두 번재 정수 입력 >> '))
op = input('연산자 입력(+,-) >> ')
result = cal(num1, num2, op)
print(f"결과 : {result}")
```

- 함수 - 매개변수 기본 값 설정 

```python
# 숫자데이터에 제곱을 구해서 되돌려주는 함수
def power_num(num, power = 2):
    return num ** power
```

```python
# 설정된 기본값으로 계산 -> 기본값으로 계산 시 power를 입력하지 않아도 됨
power_num(3)
```

```python
# 3제곱으로 변경하여 연산하고 싶을때(기본값 변경)
power_num(4, 3)
```

```python
# 사용자가 입력한 모든 숫자들의 합을 구하는 함수 정의
def all_sum(*args):
    result = 0
    for num in args: # args(튜플)의 모든 데이터를 num에 담아 연산
        result += num
    return result
```

```python
all_sum(1,2,3,4,5,6,7,8,9)
```
