- 리스트 내부 요소 접근
- 인덱싱: 무서인가를 가리킬 때 사용(1개)
- 슬라이싱: 무엇인가를 잘라낼 때 사용(여러개)

```python
# 0번 사람의 정보를 출력해줘 - 인덱싱 (데이터 접)
person[0]
```

```python
# 0번째 사람의 생일만 출력해줘
person[0][0]
```

```python
# 1번째 사람의 이르므마나 출력해줘
person[1][1]
```

```python
# 슬라이싱
# 0~1번 사람의 정보에 접근해보기
# 대상[시작인젝스:끝인덱스+1:step]

person = [['01/12','지서영'],['09/17','박기준'],['10/16','홍길동']]
person[0:2]
person[:2:2]
```

```python
# 사람의 이름만 출력해보기
# print(person[0][1])
# print(person[1][1])
# print(person[2][1])

for i in [0,1,2]:
    print(person[i][1])

# - 리스트의 내부 형태를 동일하게 맞추는게 추후 자동화 처리에 도움이 됨
```

```python
# + 더하기 연산 > 리스트의 길이가 늘어난 형태로 결고반환
# b + 2 >> 리스트와 단일 요소는 더해질 수 없음, 숫자 연산 불가
b = [1,2,3]
b = b + [2] # 복합 대입 연산자로 변경한다면> b += [2]

b
```

```python
# lst 리스트 선언
lst = ['딸기', '바나나', '수박', '체리', '포도']
print("*"*45)

# 검색 문자 입력 받기
search = input('검색할 문자를 입력하세요 >> ')    

# lst 내부에 입력을 받은 검색 문자가 있는지 확인(in)
if search in lst:    
    # 있으면, 볓번째 인덱스에 있는제(list - index())
    print(f'{search}는 리스트 {lst.index(search)}번째 인덱스에 들어있습니다.')
else:
    # ㅇㅄ으면, 없다는 문구 출력
    print(f'{search}는 리스트에 들어있지 않습니다.')
print("*"*45)
```

```python
S = "Hi, My name is Mihee"
print("*"*40)
search = input('검색할 문자를 입력하세요 : ')
if search in S:
    print(f"'{search}'는 {S.count(search)} 번 들어 있습니다.")
else:
    print(f"'{search}'는 문자열에 들어있지 않습니다.")
```
