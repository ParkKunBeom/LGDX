##### 인덱싱과 슬라이싱
- 인덱싱(indexing) : 무엇인가를 가리킨다는 의미, 인덱스를 활용해 데이터 문자 가져오기
- 슬라이싱(slicing) : 무엇인가를 잘라낸다는 의미, 인덱스를 활용해 원하는 범위만큼 가져오기

```python
s = "My name is KB"
print(s[12])
print(s[8])
print(s[-2])
```

```python
print(s[8:10])
print(s[3:7])
```

```python
day = '2025년 12월 31일의 날씨는 흐림입니다.'
print('날짜 :', day[:13])
print('날씨 :', day[-6:-4]) 
print(day[-6:])
```

```python
y= '2025'
m= 12
d= '31'
print('오늘은 %s년 입니다.'%y)
print('오늘은 %s년 %d월 입니다'%(y,m))
print('오늘은 %s년 %d월 %s일 입니다'%(y,m,d))
```

##### 문자열 포매팅(formatting)
- 문자열에 특정한 데이터를 포함시키고자 할 때 사용
  1. %기호 포매팅 : 모든 언어에서 통용되는 방법
  2. format함수 : python에서만 활용 가능
  3. f문자열 포매팅 : python에서만 활용 가능

1. %기호 포매팅



2. format함수 활용

```python
y = '2025'
m = 12
d = '31'

s = '오늘은 {}년 {}월 {}일 입니다.'.format(y,m,d)
print(s)
```

3. f문자열 포매팅

```python
y = '2025'
m = 12
d = '31'

s = f'오늘은 {y}년 {m}월 {d}일 입니다.'
print(s)
```
