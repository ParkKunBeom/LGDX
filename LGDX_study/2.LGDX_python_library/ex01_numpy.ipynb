##### 넘파이배열 인데싱, 슬라이싱
- 인덱싱: 인덱스 번호를 활용하여 요소 1개 추출
  - 변수명.[인덱스 번호]
- 슬라이싱: 범위를 주어 여러개의 요소를 추출
  - 변수명[시작인덱스 : 끝 + 1]

```python
# 인덱싱
arr = np.array([[1,2,3],[4,5,6]])
arr
```

```python
# 숫자 2 추출
print(arr[0,1])
```

```python
# [행,열]
arr[0,1]
```

```python
# 슬라이싱
# 0~9까지 갑싱 담긴 1차원 배열
# arrange : array + range -> 일정한 규칙으로 증가하는 배열 생성
# arange(시작값, 끝값+1, 증감수)
arr = np.arange(10)
arr
```

```python
# 2~5까지 출력
arr[2:6]
```

- unique한 값 골라내기
  - np.unique(배열)

```python
# unique한 item(영화) id 추출
uniqueItemId = np.unique(ratings[:,1])
uniqueItemId
```

```python
# unique한 use(사용자) id 추출
uniqueUserId = np.unique(ratings[:,0])
```

```python
len(uniqueUserId)
```

```python
uniqueUserId
```

##### 예제 2-3 각 사용자별 평점 구하기 

```python
userList = []
for userId in uniqueUserId:
    user=float(ratings[ratings[:,0]==userId, 2].mean())
    userList.append([int(userId),user])
```

```python
userList[:5]
```

```python
itemList = []
for itemId in uniqueItemId:
    item=float(ratings[ratings[:,1]==itemId, 2].mean())
    itemList.append([int(itemId),item])
```

```python
itemList[:5]
```
