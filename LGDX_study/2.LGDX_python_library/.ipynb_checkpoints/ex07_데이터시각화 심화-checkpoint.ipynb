### 요일별 사고 건 수를 시각화하기


```python
data2 = data['요일'].value_counts()
```

```python
day = data2[['월','화','수','목','금','토','일']]
```

```python
day.values
```

```python
day.index
```

```python
# data['요일'].value_counts().loc[['월','화','수','목','금','토','일']].plot(kind='bar',rot=0, xlabel='')

x = day.index
y= day.values
plt.bar(x,y)
plt.show()
```

```python
# day.plot(kind='bar',
#          width=0.3,
#          ylim=(500,620), 
#          xlabel='요일', 
#          ylabel='사고건수', 
#          title='요일별 사고건수',
#          color = 'y', 
#          edgecolor='r', 
#          rot=0,)

#그래프 색 변경
plt.bar(x,y, width=0.3, color = 'y', edgecolor='r')
#500~620사이의 데이터를 명백히 구분할 수 있게
plt.ylim(500, 620)
#그래프의 이름 설정하기
plt.title('요일별 사고건수')
#x축 이름 설정
plt.xlabel('요일', loc='left')
#y축 이름 설정하기
plt.ylabel('사고건수', labelpad=50, rotation=0)
plt.show()
```

##### 사고유형 중 차대차 사건 중 죽거나 다친 사람이 많은 발생지시도를 알아보고 시각화해보자!

```python
data.columns
```

```python
data['사고유형_대분류']
```

```python
data['사고유형_대분류'].value_counts()
```

```python
#사고유형 대분류 중 "차대차" 교통사고만 가져오기! -> boolean인덱싱활용하기!
car = data[data['사고유형_대분류']=='차대차']
car
```

```python
#car 데이터에서 각 발생지시도에서 발생한 사상자수 확인해보기!
data1 = car[['사상자수','발생지시도']]
data1
```

```python
data1['발생지시도'].value_counts()
```

```python
#groupby -> 같은 값을 하나로 묶어서 총계, 합계 결과를 가져오자!
result = data1.groupby('발생지시도').sum()
result
```

```python
#라인그래프로 보여주기!
x = result.index #발생지시도
y = result.values #사상자수
plt.plot(x,y)
plt.show()
#이런 선 그래프는 시계열 데이터에 적합하다!
```

```python
y = result["사상자수"].values
```

```python
#바차트로 그리기! -> matplotlib에서는 막대그래프 그릴 때 x,y축이 1차원으로 동일해야한다!
plt.bar(x,y)
plt.show()
```
