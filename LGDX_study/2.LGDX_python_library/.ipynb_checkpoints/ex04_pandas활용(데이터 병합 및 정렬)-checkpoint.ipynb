- org_info 파일을 읽어와 org_info 변수에 저장
  - 인코딩 : 'euc-kr'
  - 'ORG_CD' 컬럼을 인덱스로 변경

```python
org_info = pd.read_csv('./data/org-info.csv', encoding = 'euc-kr',index_col ='ORG_CD')
```

```python
org_info
```

```python
hr_info = pd.read_csv('./data/hr-info.csv', encoding = 'euc-kr')
hr_info.tail()
```

```python
# hr_info 데이터 프레임에서 'ORG_CD' 조직 코드에 따른 -> 인원수, 평균 나이, 평균근속년수, 평균연봉 확인
# groupby()
org_agg = hr_info.groupby('ORG_CD').agg({'EID' : 'count',
                                          'AGE' : 'mean',
                                          'SERVICE_YEAR' : 'mean',
                                          'SALARY' : 'mean'})
# 컬럼이름 수치에 맞게 변경
org_agg.columns = ['HEAD_COUNT', 'AGE_MEAN','SERVICE_MEAN','SALARY_MEAN']
org_agg
```

```python
pd.concat([org_info, org_agg], axis=0, join = 'inner')
```

### DataFrame 병합(합치기) - Merge
- 두개의 DataFrame 에서 공통괸 키값을 기준으로 병합
- 데이터의 정보를 기준으로 병합 -> 두개의 DataFrame에 공통으로 존재

- Merge 특징
  - Join 속성
    - inner: 교집합 병합 (기본)
    - outer: 합집합 병합
    - left: 왼쪽 데이터프레임 기준
    - right: 오른쪽 데이터프레임 기준

### Concat, Merge 차이점
- concat
  - 병합기준: 인덱스명, 컬럼명
  - 단순연결 (행추가, 열추가)
- merge
  - 병합기준: 데이터값(공통된 키값)
  - 데이터셋 간의 관계를 바탕으로 결합

### apply
- DataFrame, Series 데이터에 함수를 행or열 단위로 적용할때 사용
- df.apply(함수명, 축설정) -> 기본축은 행 우선

```python
# 년-월-일 로 구성되어있는 데이터에서 연도 데이터만 추출
date_list = [{'yyyy-mm-dd': '2003-03-13'},
            {'yyyy-mm-dd': '2017-10-20'},
            {'yyyy-mm-dd': '2022-05-28'}]
date_df = pd.DataFrame(date_list, columns=['yyyy-mm-dd'])
date_df
```

```python
data = '2003-03-13'
data
```

```python
# split('문자') 특정 기호를 중심으로 문자열을 분리 해주는 함수 -> 리스트
data.split('-')[0]
```

```python
# 사용자 정의함수 생성
def extract_year(data):
    return data.split('-')[0]
```

```python
#DataFrame에 함수 적용하기~~
# 바로 적용 X => apply 함수를 활용하여 가능
date_df['yyyy-mm-dd'].apply(extract_year)
```

```python
# date_df 에 'year' 컬럼을 추가하여 저장
date_df['year'] = date_df['yyyy-mm-dd'].apply(extract_year)
date_df
```

### apply 적용 실습
- 내용: 직원의 성과점수'PERFORMANCE_SCORE'를 기준으로 성과등급'PERFORMANCE_GRADE' 컬럼을 새로 생성
- 조건
  - 사용자 정의함수 'get_grade' 정의하여 hr_info에 적용 후 새로운 컬럼 추가
- 성과등급 기준
    | 성과 점수 | 성과 등급 |
|----------|----------|
| 4.5 이상 | S |
| 4.0 이상 | A |
| 3.0 이상 | B |
| 그 외 | C |

```python
# 사용자 정의함수 생성 'get_grade'
# 점수에 따른 등급 부여 (조건문을 통하여 결과를 도출)
def get_grade(data):
    if data>=4.5:
        return 'S'
    elif data>=4.0:
        return 'A'
    elif data>=3.0:
        return 'B'
    else:
        return 'C'
```

```python
# apply 함수를 통하여 get_grade 함수 적용
hr_info['PERFORMANCE_SCORE'].apply(get_grade)
```

```python
# 결과를 hr_ifo에 'PERFORMANCE_GRADE' 컬럼 추가
# hr_info['PERFORMANCE_GRADE']=hr_info['PERFORMANCE_SCORE'].apply(get_grade)
hr_info.insert(loc =9,column= 'PERFORMANCE_GRADE',value = hr_info['PERFORMANCE_SCORE'].apply(get_grade))
hr_info
```

1. 2015~2017년 데이터 불러오기(index_col=관서명)
2. 구분컬림이 발생건수인 경우의 총 계
    - 2015년 발생건수 총계
    - 2016년 발생건수 총계
    - 2017년 발생건수 총계
3. 증감률 계산
    - 15-16 증감률 계산
    - 16-17 증감률 계산
4. DF로 만들기

```python
crime15 = pd.read_csv('./data/crime/2015.csv', encoding='euc-kr',index_col='관서명')
crime16 = pd.read_csv('./data/crime/2016.csv', encoding='euc-kr',index_col='관서명')
crime17 = pd.read_csv('./data/crime/2017.csv', encoding='euc-kr',index_col='관서명')
```

```python
sum15 = crime15[crime15['구분']=='발생건수'].loc[:,'살인':'폭력'].sum(axis=1)
sum16 = crime16[crime16['구분']=='발생건수'].loc[:,'살인':'폭력'].sum(axis=1)
sum17 = crime17[crime17['구분']=='발생건수'].loc[:,'살인':'폭력'].sum(axis=1)
```

```python
inc1 = (sum16-sum15)/sum15*100
inc2 = (sum17-sum16)/sum16*100
```

```python
# tatal_index=crime15.index.unique()
# result_crime=pd.DataFrame()
# result_crime['2015총계']=sum15
# result_crime['2015-2016증감율']=inc1
# result_crime['2016총계']=sum16
# result_crime['2016-2017증감율']=inc2
# result_crime['2017총계']=sum17

# result_crime
```

```python
result_crime=pd.concat([sum15,inc1,sum16,inc2,sum17], axis=1, join='inner')
result_crime.columns=['2015총계','2015-2016증감율','2016총계','2016-2017증감율','2017총계']

result_crime
```
