## 3. 전처리 응용실습

### 3-1. Age

- Age 결측치를 전체 평균 나이값으로 채우기

```python
train['Age'].fillna(train['Age'].mean().astype(int))
```

- Age 결측치를 성별 평균 or 중간 값으로 채우기

```python
# 성별에 따른 나이 평균값
train.groupby('Sex')['Age'].median()
```

```python
train['Age'].fillna(train.groupby('Sex')['Age'].transform('median'), inplace=True)
```

```python
train.tail(5)
```

### 3-2 가장 빈도가 많은 값으로 Embarked 결측치 채우기

```python
train['Embarked'].value_counts().index[0]
```

```python
train['Embarked'].fillna(train['Embarked'].value_counts().index[0])
```

```python
train['Embarked'].fillna(train['Embarked'].value_counts().idxmax(), inplace=True)
```

```python
train['Embarked'].isnull().sum()
```

```python
train['Age'].isnull().sum()
```

- 나이대별 생존/사망 비율 확인

```python
train['Age'].value_counts()
```

```python
bins = [-1,9,19,29,39,49,59,69,79,89]
labels = ['10대미만','10대','20대','30대','40대','50대','60대','70대','80대']
tr_cats = pd.cut(train['Age'],bins, labels=labels)
```

```python
train.insert(loc=5,column= 'CATS',value = tr_cats)
```

```python
pd.DataFrame(train.groupby('CATS')['Survived'].value_counts(1)*100)
```

```python
pd.DataFrame(train.groupby(['Sex','CATS'])['Survived'].value_counts(1)*100)
```

```python
agez=train.groupby('CATS')['Survived'].value_counts(1)*100

# 1. 표 형태로 펼치기
# fillna(0)은 데이터가 없는 경우(예: 80대 사망자 0명)를 0으로 채워줍니다.
agez_pretty = agez.unstack().fillna(0)

# 2. 컬럼명 읽기 쉽게 변경 (0: 사망, 1: 생존)
agez_pretty.columns = ['사망 (%)', '생존 (%)']

# 3. 출력
display(agez_pretty) 
```
