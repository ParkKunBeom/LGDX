### selenium 라이브러리
- 자동화 수집 도구
- requests 라이브러리의 역할: 브라우저를 대신하여 페이지의 정보를 요청, 응답
- requests 라이브러리가 가지는 한계인 브라우저 조작 기능을 해주는 라이브러리
- 클릭, 입력, 스크롤, 화면전환 등 사라이 하는 행동을 똑같이 가능하게 해줌
- 주의: 코드 작성시 사람의 행동을 그대로 코드로 작성(명령을 내리는 순서가 중요하다)

```python
# 라이브러리 불러오기
# 컴퓨터용 브라우저 -> 데이터를 요청하고 응답받는 역할
from selenium import webdriver as wb
# 컴퓨터용 키보드(Enter, Del, End) 컴터에게 키보드를 전달하는 역할
from selenium.webdriver.common.keys import Keys
# 선택자를 구분할때 사용하는 도구(필수)
from selenium.webdriver.common.by import By
# 쉬는 시간을 주는 라이브러리 # time.sleep(2) -> 페이지 요청 후 로딩을 기다려주기 위하여 상요
import time
```

### 실습
- 네이버 메인페이지를 열어 검색어 결과 수집하기~

##### 코드 작성 순서
1. 브라우저 열기
2. 네이버 페이지로 이동하기
3. 검색창 클릭하기
4. 검색어 입력하기
5. 엔터

```python
#1. 브라우저 열기
driver = wb.Chrome()
```

```python
#2. 네이버 페이지로 이동하기
# driver 에게 요청 -> 크롬창이 열린 상태에서 명령
driver.get('https://www.naver.com/')
```

```python
#3. 검색어 입력하기 (keyword: 강아지)
# 검생창의 위치를 드라이버에게 공유
search = driver.find_element(By.ID,'query')
#4. 검색어 입력
search.send_keys('강아지')
```

```python
#5. 엔터
search.send_keys(Keys.ENTER)
```

```python
# 페이지에 있는 강아지의 종류 수집
driver.find_elements(By.CSS_SELECTOR,'a._text')[0].text
```

```python
# 페이지에 존재하는 모든 강아지 종을 출력
dogs = driver.find_elements(By.CSS_SELECTOR,'a._text')
for i in dogs:
    print(i.text)
```

- 뒤로가기

```python
# 메인페이지로 이동하기 위해서 뒤로가기
# 뒤로가기 버튼은 html 구조에 미포함 -> 웹브라우저의 기능 -> driver에게 바로 요청
driver.back()
```

```python
#'고양이' 검색
search = driver.find_element(By.ID,'query')
search.send_keys('고양이')
# search.send_keys(Keys.ENTER)
```

```python
# 돋보기 버튼
driver.find_element(By.ID,'search-btn').click()
```

```python
# 고양이 종류 수집
cats = driver.find_elements(By.CSS_SELECTOR,'a._text')
for i in cats:
    print(i.text)
```

- 창닫기

```python
driver.quit()
```

##### 코드 합치기
- 페이지를 열고 검색 후 정보수집, 창닫기

```python
driver = wb.Chrome()
driver.get('https://www.naver.com/')
#페이지 로딩 후 쉬는시간
time.sleep(1)
search = driver.find_element(By.ID,'query')
search.send_keys('강아지')
search.send_keys(Keys.ENTER)
# 쉬는시간
time.sleep(2)
dogs = driver.find_elements(By.CSS_SELECTOR,'a._text')
for i in dogs:
    print(i.text)
driver.quit()
```
