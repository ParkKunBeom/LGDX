```python
url ='https://n.news.naver.com/mnews/article/055/0001325688'
res = req.get(url)
soup = bs(res.text, 'lxml')
```

```python
# 1. 뉴스 제목 수집하기
title = soup.select('h2#title_area>span')
title[0].text
```

```python
url = 'https://news.naver.com/section/105'
res = req.get(url)
soup = bs(res.text, 'lxml')
```

```python
# 2. 뉴스의 url 주소를 수집하기 위하여 뉴스 태그정보 내의 속성 추출
news_url = soup.select('a.sa_text_title._NLOG_IMPRESSION')
# 요소의 속성을 추출하는 방법
# 요소.get('속성명')
news_url[0].get('href')
```

```python
len(news_url)
```

```python
# link_list 리스트에 순수한 링크정보만 담기!(반복문_for)
# news_url[0].get('href')
link_list = []
for i in range(len(news_url)):
    link_list.append(news_url[i].get('href'))
link_list
```

```python
# link_list2 리스트에 순수한 링크정보만 담기!(반복문_for)
link_list2 = []
for element in news_url:
    link_list2.append(element.get('href'))
link_list2
```

### 수집한 url 뉴스들의 제목, 내용을 수집하여 DataFrame으로 만들기

```python
data_list=[]
for link in link_list:
    res = req.get(link)
    soup = bs(res.text, 'lxml')

    # 뉴스 제목 수집
    title = soup.select('h2#title_area>span')[0].text
    # 뉴스 내용 수집
    content = soup.select_one('article#dic_area').text.strip()
    # data_list에 결과담기
    data_list.append([title,content,link])
```

```python
len(data_list)
```

```python
import pandas as pd
```

```python
news_df = pd.DataFrame(data_list, columns = ['뉴스제목', '내용', '링크'])
```

- data 폴더 만든 후 저장하기~
