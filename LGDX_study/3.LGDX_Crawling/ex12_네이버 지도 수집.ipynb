### 학습 목표
- 네이버 지도에서 검색을 통해 데이터를 수집할 수 있다.
- 한페이지에 iframe이 여러개인경우 데이터를 수집할 수 있다.

```python
# 라이브러리 불러오기
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys 
import time
from tqdm import tqdm
```

```python
# 웹페이지 열기
driver = wb.Chrome()
driver.get('https://map.naver.com/p/')
```

```python
# 네이버는 선택자를 난수로 설정하는 경우 있음 -> 계층선택자, 변하지않는 구분자를 사용
# 키워드 입력후 검색
search = driver.find_element(By.CSS_SELECTOR,'input.input_search')
search.send_keys('상무지구맛집'+ '\n')
```

```python
# iframe 접근
driver.switch_to.frame('searchIframe')
```

```python
# 디어리스트 연남 -> 첫번째 맛집 클릭! (iframe)
driver.find_element(By.CSS_SELECTOR, 'a.place_bluelink.YTJkH.CtW3e').click()
```

```python
# iframe1에서 iframe2 로 이동할때는 모두를 알고있는 main (원래페이지)로 이동후 가능
# iframe1 -> main -> iframe2
driver.switch_to.default_content()
```

```python
# iframe 이동
driver.switch_to.frame('entryIframe')
```

```python
# 보배반점 -> 메뉴 정보 수집을 위해 "메뉴" 탭 클릭
# 또다른 iframe
driver.find_element(By.CSS_SELECTOR, '#app-root > div > div > div.place_fixed_maintab > div > div > div > div > a:nth-child(3)').click()
```

```python
# 메뉴 내용 수집
menu = driver.find_elements(By.CSS_SELECTOR,'div.MenuContent__tit__313LA')
for i in menu:
    print(i.text)
```

```python
# 리뷰데이터 수집
driver.find_element(By.CSS_SELECTOR, '#root > div.PlaceHeader__root__qIV7\+ > header > div.slick-slider.flick > div > a:nth-child(5)').click()
```

```python
review = driver.find_elements(By.CSS_SELECTOR, 'div.pui__vn15t2 > a:nth-child(1)')
for i in review:
    print(i.text)
```

```python
driver.quit()
```
