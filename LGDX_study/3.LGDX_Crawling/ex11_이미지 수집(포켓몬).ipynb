```python
# 라이브러리 불러오기
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from tqdm import tqdm

# 파일이나 폴더를 생성,삭제해주는 파일관리시스템 라이브러리
import os #Operationg System
# 이미지의 경로(주소)를 로컬파일로 변환해주는 라이브러리
from urllib.request import urlretrieve
```

- urllib.request
> - url 주소를 열어 타 용도로 사용하는 라이브러리

```python
# 코드로 폴더 생성
# os.mkdir('경로') -> 경로를 복사할때 역슬래시는 이스케이프 코드 -> 역슬래시 2개
os.mkdir("C:\\Users\\Park_kunbeom\\Desktop\\포켓몬")
# 현재 위치에 동일한 폴더명으로 생성은 불가능!
```

```python
# 페이지 정보 요청
url = 'https://www.pokemonkorea.co.kr/pokedex#pokedex_1'
driver = wb.Chrome()
driver.get(url)
```

```python
# 이미지의 src 수집하여 img_lit에 담기
img_list=[]
img= driver.find_elements(By.CSS_SELECTOR,'img.img-fluid')
for i in img:
    img_list.append(i.get_attribute('src'))
img_list
```

### 이미지 파일 로컬에 저장하기

```python
# urlretrieve(url , '저장할 경로/파일명,확장자')
urlretrieve(img_list[0], 'C:\\Users\\Park_kunbeom\\Desktop\\포켓몬\\포켓몬.jpg')
```

```python
len(img_list)
```

```python
for i in range(len(img_list)):
    urlretrieve(img_list[i], f'C:\\Users\\Park_kunbeom\\Desktop\\포켓몬\\포켓몬도감{i}.jpg')
```

- 포켓몬 이름으로 파일 저장을 위해 이름 수집

```python
name = driver.find_elements(By.CSS_SELECTOR,'div.bx-txt>h3')
name[0].text
```

```python
# 실재 이름만 추출 -> p태그 제거
#1. 드라이버에게 p태그 위치 알려주기
p = driver.find_elements(By.CSS_SELECTOR,'div.bx-txt>h3>p')
#2. p태그 제거
for p in p:
    driver.execute_script("arguments[0].remove();", p)
```

```python
name = driver.find_elements(By.CSS_SELECTOR,'div.bx-txt>h3')
name[0].text
```

```python
name_list = [i.text for i in name]
name_list
```

```python
for i in range(len(img_list)):
    urlretrieve(img_list[i], f'C:\\Users\\Park_kunbeom\\Desktop\\포켓몬\\{name_list[i]}.jpg')
```
