#### 문제7

```python
score_list = [['김미희', 89,78,99],
              ['서대희', 67,88,77],
              ['강예진', 100,55,96]]
py=0
ja=0
db=0
print('='*45)
print('이름\tPython\tJava\tDB\t평균\t학점')
print('='*45)
for i in score_list:
    avg = int((i[1]+i[2]+i[3])/3)
    if avg>=90:
        grade='A'
    elif avg>=80:
        grade='B'
    elif avg>=70:
        grade='C'
    elif avg>=60:
        grade='D'
    else:
        grade='F'
    py+=i[1]
    ja+=i[2]
    db+=i[3]
    print(f'{i[0]}\t{i[1]}\t{i[2]}\t{i[3]}\t{avg}\t{grade}')
print('='*45)
print(f'평균\t{py//len(score_list)}\t{ja//len(score_list)}\t{db//len(score_list)}')
```

#### 문제9

```python
import pandas as pd
score_df=pd.DataFrame([['홍길동',70,80,88],
                      ['성춘향',80,90,78],
                      ['김사또',90,60,90]],
                      columns=['이름','국어','수학','과학'])
score_df
```

```python
score_mean = score_df.loc[:,'국어':'과학'].mean(axis=1)
score_df['평균']=score_mean
score_df
```

```python
score_mean2=score_df.iloc[:,1:5].mean()
score_df.loc['종합'] = score_mean2
score_df = score_df.fillna('-').round(1)
score_df
```
